```python
# Step 3: Import the SQL interface and define a helper
# This allows us to write SQL queries on the Pandas DataFrame

from pandasql import sqldf

# Define a lambda function to use for SQL queries on our df
pysqldf = lambda q: sqldf(q, globals())

# Example Query: Select top 5 highest-rated movies
query = """
SELECT title, genre, rating
FROM df
ORDER BY rating DESC
LIMIT 5;
"""

# Run the SQL query and show the result
pysqldf(query)
```

```python
# Step 4: Recommend top-rated movies from a user-specified genre

# Ask the user for their favorite genre
user_genre = input("Enter your favorite movie genre (e.g., Sci-Fi, Drama, Action, Animation, Crime): ")

# SQL query to get top 5 movies in that genre
query = f"""
SELECT title, year, rating
FROM df
WHERE genre = '{user_genre}'
ORDER BY rating DESC
LIMIT 5;
"""

# Run the query and show recommendations
print(f"\nüé¨ Top movies in {user_genre} genre:\n")
pysqldf(query)
```

```python
# Step 5: Optional - Loop for multiple recommendations

while True:
    user_genre = input("\nüé≠ Enter a movie genre (e.g., Sci-Fi, Drama, Action) or type 'exit' to quit: ")

    if user_genre.lower() == 'exit':
        print("Thanks for using the Movie Recommender! üé¨")
        break

    query = f"""
    SELECT title, year, rating
    FROM df
    WHERE genre = '{user_genre}'
    ORDER BY rating DESC
    LIMIT 5;
    """

    print(f"\nüîç Top movies in '{user_genre}' genre:\n")
    result = pysqldf(query)

    # If no result, inform the user
    if result.empty:
        print("‚ö†Ô∏è Sorry, no movies found in that genre. Try again.")
    else:
        print(result)
```
